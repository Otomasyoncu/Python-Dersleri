```python
import sys,time,threading
from PyQt5 import QtWidgets, QtGui
from PyQt5.QtCore import *
uygulama = QtWidgets.QApplication(sys.argv)

pencere01 = QtWidgets.QWidget()
pencere01.setWindowTitle("Serkan KÃ¶se")
pencere01.setGeometry(400, 150, 400, 400)
pencere01.setWindowIcon(QtGui.QIcon("logo.ico"))

grupbox01 = QtWidgets.QGroupBox(pencere01)
grupbox01.setTitle("")
grupbox01.setGeometry(150, 50, 40, 40)

grupbox02 = QtWidgets.QGroupBox(pencere01)
grupbox02.setTitle("")
grupbox02.setGeometry(150, 100, 40, 40)

grupbox03 = QtWidgets.QGroupBox(pencere01)
grupbox03.setTitle("")
grupbox03.setGeometry(150, 150, 40, 40)

def serkan():
    while 2>1:
        grupbox01.setStyleSheet("background-color: red")
        time.sleep(0.1)
        grupbox02.setStyleSheet("background-color: gray")
        time.sleep(0.1)
        grupbox03.setStyleSheet("background-color: gray")
        time.sleep(8)
        grupbox01.setStyleSheet("background-color: red")
        time.sleep(0.1)
        grupbox02.setStyleSheet("background-color: yellow")
        time.sleep(0.1)
        grupbox03.setStyleSheet("background-color: gray")
        time.sleep(2)
        grupbox01.setStyleSheet("background-color: gray")
        time.sleep(0.1)
        grupbox02.setStyleSheet("background-color: gray")
        time.sleep(0.1)
        grupbox03.setStyleSheet("background-color: green")
        time.sleep(3)

is_parcasi = threading.Thread(target=serkan)
is_parcasi.start()


def serkan2():
    i=0
    while 2>1:
        time.sleep(1)
        i = i + 1
        etiket01.setText(str(i))
        if i>11:
            i=0

is_parcasi2 = threading.Thread(target=serkan2)
is_parcasi2.start()

etiket01 = QtWidgets.QLabel(pencere01)
etiket01.setText("00")
etiket01.move(162, 20)
etiket01.setFont(QtGui.QFont("Times", 12, QtGui.QFont.Bold))

pencere01.show()
sys.exit(uygulama.exec_())



```
