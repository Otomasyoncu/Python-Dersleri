```python
import sys
from PyQt5 import QtWidgets, QtGui
from PyQt5.QtCore import *
uygulama = QtWidgets.QApplication(sys.argv)

pencere01 = QtWidgets.QWidget()
pencere01.setWindowTitle("Serkan Köse")
pencere01.setGeometry(400, 150, 400, 400)
pencere01.setWindowIcon(QtGui.QIcon("logo.ico"))

def serkan1():
    if checkbox01.isChecked() == True :
        pencere01.setWindowTitle("Tuzla")
    else:
        pencere01.setWindowTitle("Şifa")

checkbox01 = QtWidgets.QCheckBox(pencere01)
checkbox01.setChecked(True)
checkbox01.setText("1.led")
checkbox01.move(20,20)
checkbox01.stateChanged.connect(serkan1)

checkbox02 = QtWidgets.QCheckBox(pencere01)
checkbox02.setChecked(False)
checkbox02.setText("2.Led")
checkbox02.move(20,40)
checkbox02.stateChanged.connect(serkan1)

checkbox03 = QtWidgets.QCheckBox(pencere01)
checkbox03.setChecked(True)
checkbox03.setText("3.Led")
checkbox03.move(20,60)
checkbox03.stateChanged.connect(serkan1)

def serkan2():
    if checkbox01.isChecked() == True :
        grupbox01.setStyleSheet("background-color: green")
    else:
        grupbox01.setStyleSheet("background-color: red")
    if checkbox02.isChecked() == True :
        grupbox02.setStyleSheet("background-color: green")
    else:
        grupbox02.setStyleSheet("background-color: red")
    if checkbox03.isChecked() == True :
        grupbox03.setStyleSheet("background-color: green")
    else:
        grupbox03.setStyleSheet("background-color: red")

buton01 = QtWidgets.QPushButton(pencere01)
buton01.setText("Tamam")
buton01.move(20, 100)
buton01.setFixedSize(150, 30)
buton01.setFont(QtGui.QFont("Tahoma", 10, QtGui.QFont.Bold))
buton01.clicked.connect(serkan2)

grupbox01 = QtWidgets.QGroupBox(pencere01)
grupbox01.setTitle("")
grupbox01.setGeometry(100,30, 40,40)
#grupbox01.setStyleSheet("background-color: red")

grupbox02 = QtWidgets.QGroupBox(pencere01)
grupbox02.setTitle("")
grupbox02.setGeometry(150, 30, 40, 40)
#grupbox02.setStyleSheet("background-color: red")

grupbox03 = QtWidgets.QGroupBox(pencere01)
grupbox03.setTitle("")
grupbox03.setGeometry(200, 30, 40, 40)
#grupbox03.setStyleSheet("background-color: red")

pencere01.show()
sys.exit(uygulama.exec_())
```
